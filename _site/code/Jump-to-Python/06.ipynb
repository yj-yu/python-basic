## 06-1 프로그램 만들 수 있을까?
먼저, "어떤 프로그램을 짜야지" 라는 생각보다는 다른 사람들이 만든 프로그램 파일을 자세히 들여다보고 분석하는 데서 시작할 것!

예제: 구구단 프로그램

"입력"과 "출력"을 생각하라.
* 함수 이름 = GuGu
* 입력값 = 2
* 출력값 = [2, 4, 6, 8, ... , 18]
* 결과는 list로!!


````python
def GuGu(n):
    print(n)

GuGu(2)
````

````python
def GuGu(n):
    return []

GuGu(2)
````

````python
def GuGu(n):
    result = []
    result.append(n*1)
    result.append(n*2)
    result.append(n*3)
    result.append(n*4)    
    result.append(n*5)
    result.append(n*6)
    result.append(n*7)    
    result.append(n*8)    
    result.append(n*9)    
    return result

print(GuGu(2))
````

````python
def GuGu(n):
    result = []
    i = 1
    while i < 10:
        result.append(n*i)
        i += 1
    return result

print(GuGu(2))
````

## 06-2 3과 5의 배수 합하기
문제: 1000 미만의 자연수에서 3의 배수와 5의 배수의 총합을 구하라!

* 입력값: 1부터 999 자연수 찾기
* 출력값: 3의 배수와 5의 배수의 총합
* 생각해볼것:

    * 3의 배수와 5의 배수 어떻게 찾지?
   
    * 3의 배수와 5의 배수가 겹칠 떄는??

````python
n = 1
while n < 1000:
    print(n)
    n += 1
````

````python
for n in range(1,1000):
    print(n)
````

````python
# 3의 배수 찾기
for n in range(1, 1000):
    if n%3 == 0 :
        print(n)
````

````python
# 5의 배수 찾기
for n in range(1, 1000):
    if n%5 == 0:
        print(n)
````

````python
result = 0
for n in range(1, 1000):
    if n%3 ==0 or n%5 ==0:   # 중복되는 것도 고려됨!!!!
        result += n
print(result)
````

## 06-3 게시판 페이징하기
게시판물의 총 건수와 한페이지에 보여줄 게시물 수를 입력으로 주었을때, 총 페이지 수를 출력하는 프로그램.
* 함수이름: getTotalPage
* 입력값: 게시물의 총 건수(m), 한페이지에 보여줄 게시물 수(n)
* 출력값: 총 페이지 수

````python
def getTotalPage(m, n):
    return m // n + 1

print(getTotalPage(5,10))
print(getTotalPage(15,10))
print(getTotalPage(25,10))
print(getTotalPage(30,10))    # Failed result: it should be 3 instead of 3
````

````python
def getTotalPage(m, n):
    if m % n == 0:
        return m // n
    else:
        return m // n + 1
print(getTotalPage(5,10))
print(getTotalPage(15,10))
print(getTotalPage(25,10))
print(getTotalPage(30,10))    # Failed result: it should be 3 instead of 3
````

#sub_dir_search.py
import os

def search(dirname):
    try:
        filenames = os.listdir(dirname)
        for filename in filenames:
            full_filename = os.path.join(dirname, filename)
            if os.path.isdir(full_filename): # full_filename이 디렉터리인지 구분
                search(full_filename) # 디렉터리인 경우, 다시 search 함수 호출 (recursive call, 재귀호출)
            else:
                ext = os.path.splitext(full_filename)[-1]
                if ext == '.py':
                    print(full_filename)
    except PermissionError: # os.listdir 수행 시 권한이 없는 디렉터리에 접근하더라도 프로그램이 오류로 종료되지 않게 그냥 수행되도록 하기 위함.
        pass


search('/Users/hwanpyokim/Documents/106_Python/jump to python')

**os.walk**

시작 디렉터리부터 그 하위 디렉터리 검색을 쉽게 해준다. 

````python
import os
for (path, dir, files) in os.walk('/Users/hwanpyokim/Documents/106_Python/jump to python'):
    for filename in files:
        ext = os.path.splitext(filename)[-1]
        if ext == '.py':
            print('%s/%s' % (path, filename))
````
